## Data Exploration with Pandas Dataframes

```python
#create a pandas dataframe
data = pd.DataFrame(data = boston_dataset.data, columns = boston_dataset.feature_names)

#Add column with the price(target)
data['PRICE'] = boston_dataset.target
```

## Visualising Data - Histograms, Distributions, Bar Charts

```python
plt.figure(figsize = (10,6))
plt.hist(data['PRICE'], bins = 50, ec = 'black', color ='#2196f3')

plt.xlabel('Price in thousands')
plt.ylabel('Nr. of Houses')
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.distplot(data['PRICE'], bins =50, hist = True, kde = False, color = '#fbc02d')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist(data['RM'],  ec = 'black', color ='#00796b')

plt.xlabel('Average Number of Rooms')
plt.ylabel('Nr. of Houses')
plt.show()
```

```python
data['RM'].mean()
```

```python
plt.figure(figsize = (10,6))
plt.hist(data['RAD'],bins = 24, ec = 'black', color ='#7b1fa2', rwidth = 0.5) #rwidth is width of histogram bar gives space between bars

plt.xlabel('Accessibility to Highways')
plt.ylabel('Nr. of Houses')
plt.show()
```

```python
data['RAD'].value_counts()
```

```python
frequency = data['RAD'].value_counts()
#type(frequency) - Pandas Series
#frequency.index
#frequency.axes[0]

plt.figure(figsize = (10,6))

plt.xlabel('Accessibility to Highways')
plt.ylabel('Nr. of Houses')

plt.bar(frequency.index, height = frequency)
plt.show()
```

## Correlation

## $$ \rho _{XY} = corr(X,Y)$$
## $$ -1.0 \leq \rho _{XY} \leq + 1.0 $$

```python
data['PRICE'].corr(data['RM'])
```

```python
data['PRICE'].corr(data['PTRATIO'])
```

```python
data.corr() #there are different types of correlations. Pearson correlation coefficients is default. The 2 other types are kendall and spearman
```

Correlation of variable with itself is 1. And it's symmetric. A lot of datapoints for correlation is repeated.  Symmetric about diagonal

Multicollinearity is when are 2 or more predictors are highly correlated to one another. <br> --Problems: Unreliable estimates and nonsensical findings. 

But high correlation doesn't always mean multicollinearity
