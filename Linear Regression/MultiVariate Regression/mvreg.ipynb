## Correlation

## $$ \rho _{XY} = corr(X,Y)$$
## $$ -1.0 \leq \rho _{XY} \leq + 1.0 $$

```python
data['PRICE'].corr(data['RM'])
```

```python
data['PRICE'].corr(data['PTRATIO'])
```

```python
data.corr() #there are different types of correlations. Pearson correlation coefficients is default. The 2 other types are kendall and spearman
```

Correlation of variable with itself is 1. And it's symmetric. A lot of datapoints for correlation is repeated.  Symmetric about diagonal

Multicollinearity is when are 2 or more predictors are highly correlated to one another. <br> --Problems: Unreliable estimates and nonsensical findings. 

But high correlation doesn't always mean multicollinearity

```python
mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
mask
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(data.corr(), mask=mask, annot = True, annot_kws = {'size': 14})
sns.set_style('white')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

Take a look at 0.91 (tax vs rad(access to highway)) which doesn't really make that much sense. 
but we have to consider that this is pearson correlation. Pearson correlation is for continuous variables, so it is not applicable for dummy variables or RAD (discrete variable) as they are non-continous variables. 

One more limitation we have to consider is that correlation does not imply causation. 

Another limitation they only check for linear relationships.  -- Anscombe's Quartet

Identify strong correlations, simplify by excluding irrelevant data(check if there are any by testing them out), and test for multicollinearity.

```python
nox_dis_corr = round(data['NOX'].corr(data['DIS']),3)

plt.figure(figsize = (9,6))
plt.scatter(x=data['DIS'], y = data['NOX'], alpha = 0.6, s= 80, color = 'indigo')

plt.title(f'DIS vs NOX (Correlation {nox_dis_corr})', fontsize = 14)
plt.xlabel('DIS - Distance from employment', fontsize = 14)
plt.ylabel('NOX - Nitric Oxide Pollution', fontsize = 14)
plt.show()
```

```python
sns.set()
sns.set_context('notebook') #fontsize of labels
sns.set_style('whitegrid')
sns.jointplot(x = data['DIS'], y = data['NOX'], size = 7, color = 'indigo', joint_kws={'alpha': 0.5})
plt.show()
```

```python
sns.set()
sns.set_context('notebook') #fontsize of labels
sns.set_style('whitegrid')
sns.jointplot(x = data['DIS'], y = data['NOX'], kind = 'hex', size = 7, color = 'blue')
plt.show()
```
